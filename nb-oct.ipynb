```python
import pandas as pd
df = pd.read_csv('Presupuesto Mayo 2025.csv')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(19, 6))
ax = sns.barplot(
    data=df,
    x='E.S',
    y='Ppto',
    estimator=sum,  # Puedes cambiar a 'mean' para el promedio
    ci=None,       # Desactiva intervalos de confianza
    palette="viridis"
)

plt.xticks(rotation=65, ha='right') 
# Personalizar
plt.title('Presupuesto Total por sucursal', fontsize=14)
plt.xlabel('Categoría E.S', fontsize=12)
plt.ylabel('Presupuesto Total', fontsize=12)
ax.bar_label(ax.containers[0], fmt='%.0f', label_type='edge')  # Mostrar valores

plt.tight_layout()
plt.show()
```

```python
# Agrupar y sumar por zona
grouped = df.groupby('Zona').agg({'Ppto': 'sum', 'Real': 'sum'}).reset_index()

# Reorganizar datos para seaborn (formato largo)
melted_df = pd.melt(
    grouped,
    id_vars='Zona',
    value_vars=['Ppto', 'Real'],
    var_name='Tipo',
    value_name='Monto'
)

# Configuración del gráfico
plt.figure(figsize=(12, 7))
sns.set_theme(style="whitegrid")

# Crear gráfico de barras agrupadas
ax = sns.barplot(
    data=melted_df,
    x='Zona',
    y='Monto',
    hue='Tipo',
    palette={"Ppto": "#1f77b4", "Real": "#ff7f0e"},
    alpha=0.9
)

# Personalización
plt.title('Comparación de Presupuesto vs Real por Zona', fontsize=16, pad=15)
plt.xlabel('Zona', fontsize=12)
plt.ylabel('Monto Total', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Añadir etiquetas con los valores
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,.0f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center',
        xytext=(0, 5),
        textcoords='offset points',
        fontsize=9
    )

# Añadir línea de diferencia
for i, zona in enumerate(grouped['Zona']):
    ppto = grouped.loc[grouped['Zona'] == zona, 'Ppto'].values[0]
    real = grouped.loc[grouped['Zona'] == zona, 'Real'].values[0]
    plt.plot(
        [i - 0.2, i + 0.2],
        [ppto, ppto],
        color='black',
        linestyle='-',
        linewidth=1
    )
    plt.plot(
        [i - 0.2, i + 0.2],
        [real, real],
        color='black',
        linestyle='-',
        linewidth=1
    )
    plt.plot(
        [i, i],
        [ppto, real],
        color='red',
        linestyle='--',
        linewidth=1.5,
        alpha=0.7
    )

# Ajustar leyenda
plt.legend(title='Tipo', loc='upper right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python


# Agrupar y sumar por categoría E.S
grouped = df.groupby('E.S').agg({'Ppto': 'sum', 'Real': 'sum'}).reset_index()

# Reorganizar datos para seaborn (formato largo)
melted_df = pd.melt(
    grouped,
    id_vars='E.S',
    value_vars=['Ppto', 'Real'],
    var_name='Tipo',
    value_name='Monto'
)

# Configuración del gráfico
plt.figure(figsize=(28, 18))
sns.set_theme(style="whitegrid")

# Crear gráfico de barras agrupadas
ax = sns.barplot(
    data=melted_df,
    x='E.S',
    y='Monto',
    hue='Tipo',
    palette={"Ppto": "#3498db", "Real": "#e74c3c"},
    alpha=0.95,
    edgecolor='black',
    linewidth=1
)

# Personalización avanzada
plt.title('Comparación Presupuesto vs Real por scursal', fontsize=16, pad=15)
plt.xlabel('Categoría E.S', fontsize=12)
plt.ylabel('Monto Total', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=10)

# Añadir etiquetas con los valores
for container in ax.containers:
    ax.bar_label(
        container, 
        fmt='%.0f', 
        padding=3,
        fontsize=10,
        fontweight='bold'
    )

# Añadir líneas de diferencia
for i, categoria in enumerate(grouped['E.S']):
    ppto = grouped.loc[grouped['E.S'] == categoria, 'Ppto'].values[0]
    real = grouped.loc[grouped['E.S'] == categoria, 'Real'].values[0]
    
    # Línea que conecta Ppto con Real
    plt.plot(
        [i, i],
        [real, ppto],
        color='black',
        linestyle='--',
        linewidth=1.5,
        alpha=0.7
    )
    
    # Marcador en el punto medio
    mid_point = (ppto + real) / 2
    plt.plot(
        i, mid_point, 
        marker='o',
        markersize=8,
        color='#2c3e50',
        alpha=0.7
    )
    
    # Texto de diferencia
    diff = ppto - real
    plt.text(
        i + 0.35,
        mid_point,
        f'Δ: {diff:.0f}',
        fontsize=10,
        fontweight='bold',
        verticalalignment='center',
        color='#2c3e50'
    )

# Mejorar leyenda
plt.legend(
    title='Tipo de Monto', 
    loc='upper right', 
    title_fontsize=12,
    fontsize=11,
    frameon=True,
    framealpha=0.9
)

# Añadir grid y ajustar layout
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()

# Mostrar porcentaje de utilización
plt.figure(figsize=(15, 6))
grouped['Utilización'] = grouped['Real'] / grouped['Ppto'] * 100

ax2 = sns.barplot(
    data=grouped,
    x='E.S',
    y='Utilización',
    palette='Blues_r',
    edgecolor='black'
)

# Añadir valores de porcentaje
for i, row in grouped.iterrows():
    ax2.text(
        i,
        row['Utilización'] + 2,
        f"{row['Utilización']:.1f}%",
        ha='center',
        fontsize=11,
        fontweight='bold'
    )

plt.title('Porcentaje de Utilización del Presupuesto por Categoría E.S', fontsize=14)
plt.axhline(y=100, color='r', linestyle='--', alpha=0.7)
plt.ylabel('Porcentaje (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
